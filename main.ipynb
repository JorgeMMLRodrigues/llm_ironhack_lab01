```python

client = InferenceClient(
    provider="novita",
    api_key=os.environ["HF_TOKEN"],
)

# Grab the function object
fn = client.chat.completions.create

# Print its signature
print("Signature:\n", inspect.signature(fn), "\n")

# List each parameter and its default value
print("Parameters:")
for name, param in inspect.signature(fn).parameters.items():
    print(f"  • {name}: default={param.default!r}")
```

```python

resp = client.chat.completions.create(
    model="meta-llama/Llama-3.1-8B-Instruct",
    messages=[{"role":"user","content":"Explain quantum entanglement in simple terms."}],
    temperature=0.3,           # low randomness
    top_p=0.9,                 # only sample from top 90% probability mass
    max_tokens=128,            # don’t go over 128 tokens
    presence_penalty=0.2,      # small push for novel words
    frequency_penalty=0.1      # small push against repeating
)

print(resp.choices[0].message.content)
```

## Recipe Creator with Available ingredients

```python
def generate_recipe(
    ingredients,
    temperature=0.7,
    top_p=0.9,
    max_tokens=500,
    presence_penalty=0.1,
    frequency_penalty=0.1
):
    prompt = (
        "You are a world-class chef.\n"
        f"Create a recipe using only these ingredients:\n"
        f"{', '.join(ingredients)}\n\n"
        "Step-by-step instructions:\n"
        f"Be concise and clear to minimize token usage.\n"
    )
    resp = client.chat.completions.create(
        model="meta-llama/Llama-3.1-8B-Instruct",
        messages=[{"role":"user", "content":prompt}],
        temperature=temperature,
        top_p=top_p,
        max_tokens=max_tokens,
        presence_penalty=presence_penalty,
        frequency_penalty=frequency_penalty
    )
    return resp.choices[0].message.content.strip()

# quick test
print(generate_recipe(["eggs","flour","milk","sugar"]))
```

```python

grid = [
    {"temperature": 0.3, "top_p": 0.8},
    {"temperature": 0.7, "top_p": 0.9},
    {"temperature": 1.0, "top_p": 1.0},
]

for params in grid:
    print(">>>", params)
    print(generate_recipe(["eggs","flour","milk","sugar"], **params), "\n")
```

```python
test_ingredient_sets = [
    ["potato","onion","carrot"],
    ["chicken","lemon","rosemary"],
    ["tomato","basil","mozzarella"]
]

for ings in test_ingredient_sets:
    print(">>> Ingredients:", ings)
    # pass the list as the first arg, and only use named params for the API knobs
    print(generate_recipe(ings, temperature=0.7, top_p=0.9, max_tokens=500), "\n")
```
